The first dataset is on the World Happiness Report in 2016 to answer research questions based on a survey of the state of global happiness. #### The first dataset attribute information:
- <b>Country</b> : Name of the country.(nominal)
- <b>Region</b> : Region the country belongs to.(nominal)
- <b>Happiness Rank</b> : Rank of the country based on the Happiness Score.(numeric)
- <b>Happiness Score</b> : A metric measured in 2016 by asking the sampled people the question: "How would you rate  your happiness on a scale of 0 to 10 where 10 is the happiest."(numeric)
- <b>Lower Confidence Interval</b> : (numeric)
- <b>Upper Confidence Interval</b> : (numeric)
- <b>Economy</b> : The extent to which GDP contributes to the calculation of the Happiness Score.(numeric)
- <b>Family</b> : The extent to which Family contributes to the calculation of the Happiness Score.(numeric)
- <b>Health</b> : The extent to which Life expectancy contributed to the calculation of the Happiness Score.(numeric)
- <b>Freedom</b> : The extent to which Freedom contributed to the calculation of the Happiness Score.(numeric)
- <b>Trust</b> : The extent to which Perception of Corruption contributes to Happiness Score.(numeric)
- <b>Generosity</b> : The extent to which Generosity contributed to the calculation of the Happiness Score.(numeric)
- <b>Dystopia</b> : The extent to which Dystopia Residual contributed to the calculation of the Happiness Score. Each country is compared against a hypothetical nation that represents the lowest national averages for each key variable and is, along with residual error, used as a regression benchmark(numeric)

#### Explore the age group with different gender group

Since the dataset variables are only include country, sex, age group, we will use sex values data to look at suicide data for each age group in the world.

Before we do it, we will drag the data with sex value male  and female.

```python
# get only sex has Male and Female data
index_BothSexes = df_suicide_long[(df_suicide_long['Sex'] == 'Both sexes')].index
# Delete these row indexes from dataFrame
df_MF_suicide = df_suicide_long.drop(index_BothSexes)
df_MF_suicide
```

#### __Aggregation__

For male and female, we will aggregate the Suicide number over the world and do the comparison  between gender and group further.

```python
# aggregate data with sum method
df_suicide_sum = df_MF_suicide.groupby(['Sex','Age']).agg({'Suicide_num': 'sum'}).reset_index()
df_suicide_sum
```

Next, we will look at the numerical variables value description of both sexes dataset roughly because we're going to look at the data as a whole and compare suicide rates across countries. 

Before doing the exploration, we will first add a summary column that shows the total suicide number for each row.

```python
# get only sex has Both sexes data
index_MFSexes = df_suicide[(df_suicide['Sex'] == 'Male') | (df_suicide['Sex'] == 'Female')].index
# Delete these row indexes from dataFrame
df_BothSex_suicide = df_suicide.drop(index_MFSexes)

df_BothSex_suicide['Total_suicide_number'] = df_BothSex_suicide[list(df_BothSex_suicide.columns)].sum(axis=1)
df_BothSex_suicide.head(3)
```

```python
df_BothSex_suicide.describe().T
```

## 1. What attributes could be the factors that influence the happiness score?

As we mentioned before, we may need to use the heatmap to confirm the correlation attributes that we chose which are Health (Life Expectancy), Economy (GDP per Capita), Freedom, Family, Dystopia Residual.

```python
# correlation heat map
# Using Pearson Correlation
plt.figure(figsize=(12,10))

corr = df_happiness_sucide[['Happiness Score','Health (Life Expectancy)', 
                            'Economy (GDP per Capita)', 'Freedom', 
                            'Family', 'Dystopia Residual',
                            'Trust (Government Corruption)','Generosity']].corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,
        annot=True, cmap=plt.cm.Reds)
```

The heat map scores indicate that our reference is correct and that there is no relatively high correlation between Freedom and Happiness Scores.

Therefore, based on the scatter regression plot and heat map score, we determine that  'Health (Life Expectancy)', 'Economy (GDP per Capita)', 'Family' have a relatively high correlation with happiness.

## 2. Based on GDP per capita, does a developing country have a higher suicide rate or a developed country has a higher suicide rate?

```python
# check the developed and developing country count.
df_happiness_sucide.Status.value_counts()
```

Since the number of developed country is only 29, which is far below the number of developing countries. we will use the line plot to explore our question.

```python
fig, ax = plt.subplots(figsize=(12,7))
ax = sns.lineplot(ax = ax, data=df_happiness_sucide, x='Economy (GDP per Capita)', 
                  y="Total_suicide_number", hue = 'Status', style="Status", markers=True, dashes=False)
ax.legend(labels = status_name_mapping)
ax.set_title('Compare the Number of Suicides in Developing and Developed Countries based on Economic')
```

Based on the line plot graph, since developed countries are in the minority, we can only start with per capita GDP greater than 1.1. With the increase of per capita GDP, the suicide rate in developed countries has no obvious trend change, but the suicide rate in developing countries has significantly decreased and the trend is stable and lower than that in developed countries.
As a result, suicide rates in developing countries with higher GDP per head are lower than in developed countries.
